# Analysing Used Car Data from Ebay :

```python
Over 371528 used cars scraped from EBay. Those fields are

•	Date Crawled : when this ad was first crawled, all field-values are taken from this date
•	Name : "Name" of the car
•	Seller : private or dealer
•	Offer Type
•	Price : the price on the ad to sell the car
•	Abtest
•	Vehicle Type
•	Year Of Registration : at which year the car was first registered
•	Gearbox
•	Power PS : Power of the car in PS
•	Model
•	Kilometer : How many kilometers the car has driven
•	Month Of Registration : at which month the car was first registered
•	Fuel Type
•	Brand
•	Not Repaired Damage : if the car has a damage which is not repaired yet
•	Date Created : The date for which the ad at eBay was created
•	Nr Of Pictures: Number of pictures in the ad 
•	Postal Code
•	Last Seen Online : when the crawler saw this ad last online
```

```python
This dataset contains the data of used cars in germany. Here we are going to work on this dataset of used cars listed from Ebay and this data was scraped from Kaggle.
The main objetive of this dataset is to train a linear regression model in order to predict the price of a used car.
Here we are going to :
1) Analyze the data. 
2) Clean the dataset step by step. 
3) Visualize the data using various plots like histogram,box plot,bar plot etc. 
4) Applying various algorithms i.e. regression model or regressors to find the maximum accuracy of the model.
```

```python
In this dataset, the following changes I have made:
1) Replace the missing values of categorical column with my name i.e. 'Himanshu'.
2) Changing the variables of notRepairedDamage column i.e., "ja" with "yes" and "nein" with "no" and replacing the missing values with "nein" as "nein" is the mean.
3) Replacing the missing values in gearbox column with "automatic" and "manuell".
4) Calculating the date and time difference from the dateCreated and lastSeen column so as to determine the time from when the ad is online. 
```

# Finding outliers and replacing them by LTV & HTV

```python
# Detecting outliers in car1 dataset
def outlier_detect(car1):
    for i in car1.describe().columns:
        Q1=car1.describe().at['25%',i]
        Q3=car1.describe().at['75%',i]
        IQR=Q3-Q1
        LTV=Q1-1.5*IQR
        UTV=Q3+1.5*IQR
        car1[i]=car1[i].mask(car1[i]<LTV,LTV) # Replacing outliers by LTV which are less than LTV
        car1[i]=car1[i].mask(car1[i]>UTV,UTV) # Replacing outliers by UTV which are less than UTV
    return car1
```

```python
car1=outlier_detect(car1)
```
